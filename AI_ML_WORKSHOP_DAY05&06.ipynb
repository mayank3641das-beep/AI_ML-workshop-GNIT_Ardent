```python
def convert(x):
  if isinstance(x,str):
    x=x.strip()
    if "-" in x:
      try:
        a,b=x.split("-")
        return (float(a)+float(b))/2
      except:
        return float(x)
    try:
      return float(x)
    except:
      return np.nan
  return np.nan
```

```python
df['total_sqft']=df['total_sqft'].apply(convert)
```
