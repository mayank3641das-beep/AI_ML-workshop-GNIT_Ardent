```python
# step:1 = import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# optional display setting
pd.set_option("display.max_columns", None)
```

```python
# step:2 = Data collection
#creating a sample student dataset

data = {
    "student_ID":[101,102,103,104],
    "Name":["kaustav","prince","mayank","prithvish"],
    "Gender":["male","male","male","male"],
    "math":[99,92,90,90],
    "science":[99,99,98,99],
    "english":[89,90,99,99],
    "attendence":[67,67,67,41],
    "study_hours":[3.5,5.6,8.9,1.4]
}
df = pd.DataFrame(data)
print("=======ORIGINAL DATASET=======")
print(df)
```

```python
#Step 3: basic information
#=========================

print("\n===== DATASET INFO =====")
print(df.info())

print("\n===== FIRST 5 ROWS =====")
print(df.head())

print("\n===== LAST 5 ROWS =====")
print(df.tail())

print("\n===== SHAPE OF DATASET ===== ")
print("Rows,columns:", df.shape)

print("\n===== COLUMN NAMES =====")
print(df.columns)
```

```python
# step 4 : check missing values
print("\n===== MISSING VALUES =====")
print(df.isnull().sum())
```

```python
# step 5 : data processing
#filling missing numeric values with column mean

df['math'].fillna(df['math'].mean(), inplace=True)
df['science'].fillna(df['science'].mean(), inplace=True)
df['english'].fillna(df['english'].mean(), inplace=True)
df['attendence'].fillna(df['attendence'].mean(), inplace=True)

print('\n===== DATASET AFTER MISSING VALUES =====')
print(df)

print('\n===== MISSING VALUES AFTER CLEANING =====')
print(df.isnull().sum())
```

```python
# step 6 : add new columns
#total and average marks

df['total']=df['math']+df['science']+df['english']
df['average']=df['total']/3

#grade calculation

def assign_grade(average):
    if average>=90:
        return 'A'
    elif average>=80:
        return 'B'
    elif average>=70:
        return 'C'
    elif average>=60:
        return 'D'
    else:
        return 'F'
df["grade"]= df["average"].apply(assign_grade)

print('\n===== DATASET WITH TOTAL AVERAGE GRADE =====')
print(df)
```

```python
# step:7 = Descriptive statistics
#================================

print("\n======= DESCRIPTIVE STATISTICS ======")
print(df.describe())
```

```python
# step:8 = Top performer

topper = df.loc[df["average"].idxmax()]

print("\n====== TOPPER STUDENT ======")
print("Name", topper["Name"])
print("average marks:" , round(topper["average"],2))
print("grade:", topper["grade"])
```

```python
# step:9 = visualization by pie chart
#====================================
grade_counts = df["grade"].value_counts()

plt.figure(figsize=(6,7))
plt.pie(grade_counts, labels=grade_counts.index, autopct="%1.1f%%" , startangle=90)
plt.title("Grade Distribution")
plt.show()
```

```python
# step 9 : seaborn scatter plot
#==============================
plt.figure(figsize=(6,7))
sns.scatterplot( x="study_hours", y="average", hue="Gender" ,data=df, s =100, color="greens")
plt.title("Study Hours vs Average Marks")
plt.show()
```

```python
#step 10 : Final output
#======================

print("\n===== FINAL DATASET =====")
print(df)
```
