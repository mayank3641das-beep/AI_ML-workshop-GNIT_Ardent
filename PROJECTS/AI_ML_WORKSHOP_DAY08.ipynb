```python
#STEP1

import pandas as pd
import numpy as np

import re

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# STEP 2: CREATE A SMALL BEGINNER DATASET
# ---------------------------------------------------------

# We are creating the dataset manually so it can run directly in Colab
# label = spam or ham
# message = the text message

data = {
    "label": [
        "ham", "spam", "ham", "spam", "ham", "spam", "ham", "spam", "ham", "spam",
        "ham", "spam", "ham", "spam", "ham", "spam", "ham", "spam", "ham", "spam"
    ],
    "message": [
        "Hey, are we meeting today?",
        "Congratulations! You won a free iPhone. Click now!",
        "Please send me the notes of today class",
        "Win cash prize now by clicking this link",
        "Can you call me after lunch?",
        "Limited offer! Buy now and get 50 percent discount",
        "I am on the way to college",
        "You have been selected for a free vacation",
        "Let's study together in the evening",
        "Claim your lottery reward today",
        "Can you help me with Python homework?",
        "Urgent! Your account has won a huge reward",
        "Where are you now?",
        "Free entry in contest, click here now",
        "I will reach home by 8 PM",
        "Get a free gift voucher instantly",
        "Did you complete the assignment?",
        "Exclusive deal only for you, act fast",
        "Let's go for tea after class",
        "You are a lucky winner, claim your prize"
    ]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Show first 5 rows
print("First 5 rows of dataset:")
display(df.head())
```

```python
# STEP 3: TEXT CLEANING
# ---------------------------------------------------------

df['cleaned_message'] = df['message'].apply(clean_text)

print("First 5 rows with cleaned messages:")
display(df[['message', 'cleaned_message']].head())
```

```python
#step 4 : text cleaning function

def clean_text(text):

  text=text.lower()

  text=re.sub(r'[^a-zA-Z0-9\s]', '', text)

  text = re.sub(r'\s+', ' ', text)

  return text
```

```python
#step 5 : LABEL ENCODING

label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['label'])


print("\nOriginal and encoder labels:")
print(df[['label', 'label_encoded']])

```

```python
#step6 : separate features and target

x = df['message']
y = df['label_encoded']

print("\nFeatures (x):")
print(x.head())

print("\nTarget (y):")
print(y.head())

```

```python
#STEP 7: TRAIN-TEST SPLIT
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#step8

vectorizer = CountVectorizer()
x_train_vect = vectorizer.fit_transform(x_train)
x_test_vect = vectorizer.transform(x_test)

print("Shape of x_train_vect:", x_train_vect.shape)
print("Shape of x_test_vect:", x_test_vect.shape)
```

```python
#step9 : make predictions

model = MultinomialNB()
model.fit(x_train_vect, y_train)

y_pred = model.predict(x_test_vect)

print("Predictions (y_pred):")
print(y_pred)
```

```python
#step10

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Step 11: Eliminate the model

# Delete the model and vectorizer objects to free up memory
# Note: This will make the 'model' and 'vectorizer' variables unavailable for further use.
del model
del vectorizer

print("Trained model and vectorizer have been eliminated from memory.")
```

```python
#step 12: TEST WITH NEW CUSTOM MESSAGES

new_messages = [
              "congratulations you won a free gift",
              "can you send me your war thunder password",
              "claim your war thunder op's"
]


x_train_vect = vectorizer.fit_transform(x_train)
# x_test_vect = vectorizer.transform(x_test) # Not strictly needed for new message prediction

# Re-create and train the model (equivalent to original step 9)
model = MultinomialNB()
model.fit(x_train_vect, y_train)

# The clean_text function is already defined in the notebook.

# Define the prediction function (as it was before 'elimination' in Step 11)
def predict_message(model, vectorizer, message):
    cleaned_message = clean_text(message)
    vect_message = vectorizer.transform([cleaned_message])
    prediction = model.predict(vect_message)
    return "spam" if prediction[0] == 1 else "ham"

print("\nPredictions for new messages:")
for msg in new_messages:
    prediction = predict_message(model, vectorizer, msg)
    print(f"'{msg}' is: {prediction}")
```
